# 1. Prepare your files
 
First we need to your Arvest user email and password and the path of the folder were you have stock your markdown files and of the folder where you want to load your manifest files before upload there to Arverst
For each markdown files in the markdown folder, this script will create a separate json file and upload it on Arvest.

```python
# Your Arvest user email and password

ARVEST_MAIL = "mail"
ARVEST_PASSWORD = "password"

# The folder's path were you want to load your manifest

MARKDOWN_PATH = os.path.join(os.getcwd(),"your markdown folder")
MANIFEST_PATH = os.path.join(os.getcwd(),"your manifest folder")
```

# 3. Launch the script

```python
nbr = os.listdir(MARKDOWN_PATH)
md_count = 0
for i in nbr :
    md_count += 1
print(f"you have {md_count} markdown files in your folder")


num_fichier = 1

for i in range(md_count):
    
    MARKDOWN_FILE = os.path.join(MARKDOWN_PATH, f"Manifest_{num_fichier}.md")

    md_to_manifest(MARKDOWN_FILE, MANIFEST_PATH, ARVEST_MAIL, ARVEST_PASSWORD)

    metadata_update(MARKDOWN_FILE, MANIFEST_PATH, ARVEST_MAIL, ARVEST_PASSWORD)
    
    md_to_anotation(MARKDOWN_FILE, MANIFEST_PATH, ARVEST_MAIL, ARVEST_PASSWORD)


    num_fichier += 1
```
